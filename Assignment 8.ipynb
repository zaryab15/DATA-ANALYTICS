```python
import csv
```

```python
open_file = open('student data.csv' , mode = 'r')
open_file
```

```python
read = csv.reader(open_file)
read
```

```python
data = list(read)
data
```

```python
header = data[0]
header
```

```python
std_data = data[1:]
std_data
```

Average of Physics Marks

```python
phy_marks = []
for i in range (0,len(std_data)-1):
    phy_marks.append(float(std_data[i][3]))
```

```python
phy_marks
```

```python
avg_physics = sum(phy_marks)/len(phy_marks)
avg_physics
```

Average Maths Marks

```python
math_marks = []
for i in range (0,len(std_data)-1):
    math_marks.append(float(std_data[i][4]))
```

```python
math_marks
```

```python
avg_math = sum(math_marks)/len(math_marks)
avg_math
               
```

```python
for i in range (0,len(std_data)-1):
    percentage = round(((float(std_data[i][2]) + float(std_data[i][3]) + float(std_data[i][4]))/300*100),1)
    std_data[i].append(percentage)
```

```python
std_data
```

```python
data[0].append('Percentage')
```

```python
data
```

```python
for i in range (0,len(std_data)-1):
    if std_data[i][5] > 80:
        std_data[i].append("A+")
    elif std_data[i][5] < 80 and std_data[i][5] > 70:
        std_data[i].append("A")
    elif std_data[i][5] < 70 and std_data[i][5] > 60:
        std_data[i].append("B")
    elif std_data[i][5] < 60 and std_data[i][5] > 50:
        std_data[i].append("C")
    else:
        std_data[i].append("Fail")
```
