## ***16. Create A Boxplot Of The "Income" Column Against The "Age" Column, Segmented By Gender***

```python
plt.figure(figsize = (10,5))
sns.boxplot(x='income',y = 'age' , hue = 'gender', data = df , palette = 'viridis')
plt.title('Boxplot of Income vs Age')
plt.xlabel('Income')
plt.ylabel('Age')
plt.show()
```
